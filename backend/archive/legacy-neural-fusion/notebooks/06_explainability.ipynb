# Transparent explainability
Instead of untested SHAP infrastructure, show the largest normalized cohort deviations. This matches the production API's contributing signals and remains auditable.

```python
import json
from pathlib import Path
artifact=json.loads((Path('../models/industry4_neural_fusion_model.json')).read_text())
demo=[80,82,78,85,100,45,83,81,82,79,80,75,74,72,76,75,80,78,79]
signals=sorted(zip(artifact['features'], [abs((value-mean)/scale) for value,mean,scale in zip(demo, artifact['normalization']['mean'], artifact['normalization']['scale'])]), key=lambda item:item[1], reverse=True)[:3]
print({'top_normalized_deviations': [(name, round(score, 3)) for name,score in signals], 'caveat': 'Deviation is a descriptive signal, not causal attribution.'})
```

Full SHAP is intentionally deferred: the ZivaBasa dependency stack documents Keras/SHAP compatibility issues, and feature deviation is more reliable for this deadline.
