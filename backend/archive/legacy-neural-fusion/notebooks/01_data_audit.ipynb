# ZivaDzidzo data audit
Run this before training. It records schema, missingness, class balance, and the production privacy boundary.

```python
import csv
from collections import Counter
from ml_config import INDUSTRY4_FILE, READINESS_TARGET, GAP_TARGET
rows = list(csv.DictReader(INDUSTRY4_FILE.open(encoding='utf-8')))
print({'rows': len(rows), 'columns': len(rows[0]), 'sample_columns': list(rows[0])[:8]})
```

```python
classes = Counter(row[READINESS_TARGET] for row in rows)
gaps = [float(row[GAP_TARGET]) for row in rows]
print({'readiness_class_counts': dict(classes), 'skill_gap_min': min(gaps), 'skill_gap_mean': round(sum(gaps)/len(gaps), 3), 'skill_gap_max': max(gaps)})
```
